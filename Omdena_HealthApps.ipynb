## Health Risk Factors

Scraping data from Google play store for apps and their ratings, name, no. of downloads, 

```python
!pip install google_play_scraper
from google_play_scraper import search
```

```python
q = ['General mental health','Meditation','EFT tapping','Stress management','Depression','Anxiety','Anger management','Suicide prevention','Addiction','Therapy','Sleep issues','OCD','Eating disorders','Self-diagnosis','Malaria prevention','Stroke prevention','Heart attack prevention','Medicine reminders','Back pain','Physical therapy','Chronic pain','Cancer','Skin care','Diabetes','Sexually transmitted diseases','Dental health','Eye health','Pregnancy','Pediatric care','Geriatric care','Preventative care','Alzheimer','Vaccination tracker','First aid','Driver safety','Emergency','General nutrition','Food tracking','Water intake tracking','Healthy recipes','Weight loss','Food allergies','Workout tracking','Yoga','Weight training','Physical therapy','Stretching']
```

```python
res = {}

for query in q:
  res[query] = []
  results = search(query)

  for ob in results:
    r = {}
    r['Name'] = ob['title']
    r['Icon'] = ob['icon']
    r['Description'] = ob['description']
    r['Score'] = ob['score']
    r['Genre'] = ob['genre']
    r['Currency'] = ob['currency']
    r['Price'] = ob['price']
    r['Free'] = ob['free']
    res[query].append(r)
```
